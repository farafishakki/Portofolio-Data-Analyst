```python
X = df_2023.drop(["Country name","year"], axis=1)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
from sklearn.cluster import KMeans
random_seed = 789

ssd = []
for k in range(2,30):
    model = KMeans(n_clusters=k, random_state=random_seed)
    model.fit(X_scaled)
    ssd.append(model.inertia_)

plt.figure(figsize=(12,8), dpi=200)
plt.plot(range(2,30), ssd, marker="o")
plt.xlabel("k")
plt.ylabel("Sum of squared distance")
plt.show()
```

```python
df_ssd = pd.DataFrame(ssd, index=range(2,30), columns=["sum_of_square_distance"])
df_ssd
```

```python
df_ssd.diff().plot(kind="bar")
```

```python
df_ssd.diff()
```

```python
from sklearn.metrics import silhouette_score

silhouette_coefficients = []

for k in range(2,30):
    model = KMeans(n_clusters=k, random_state=random_seed)
    model.fit(X_scaled)
    score = silhouette_score(X_scaled, model.labels_)
    silhouette_coefficients.append(score)
    
plt.figure(figsize=(12,8), dpi=200)
plt.plot(range(2,30), silhouette_coefficients, marker="o")
plt.xlabel("k")
plt.ylabel("Silhouette Coefficient")
plt.show()
```

```python
silhouette_coefficients
```

```python
model = KMeans(n_clusters=3, random_state=random_seed)
model.fit(X_scaled)

model.inertia_
```

```python
model.labels_
```

```python
df_2023["Cluster_3"] = model.labels_
df_2023
```

```python
low_wellbeing_countries = df_2023[df_2023["Cluster_3"] == 0].agg("mean", numeric_only=True)
```

```python
high_wellbeing_countries = df_2023[df_2023["Cluster_3"] == 1].agg("mean", numeric_only=True)
```

```python
medium_wellbeing_countries = df_2023[df_2023["Cluster_3"] == 2].agg("mean", numeric_only=True)
```

```python
Clustering_countries = pd.concat([low_wellbeing_countries,medium_wellbeing_countries, high_wellbeing_countries],axis=1)
Clustering_countries = Clustering_countries.rename(columns={0:"low_wellbeing_countries",1:"medium_wellbeing_countries",2:"high_wellbeing_countries"})
```

```python
Clustering_countries = Clustering_countries.rename(columns={0:"low_wellbeing_countries",1:"medium_wellbeing_countries",2:"high_wellbeing_countries"})
```

```python
Clustering_countries = Clustering_countries.drop(index=["year","Cluster_3"])
Clustering_countries
```

```python
from scipy.stats import shapiro
stats = Clustering_countries.apply(lambda column: shapiro(column))
stats = stats.T
stats = stats.rename(columns={0:"statistic_score", 1:"p_value"})
stats
```

```python
from scipy.stats import kruskal
statistic, p_value = kruskal(Clustering_countries["low_wellbeing_countries"],Clustering_countries["medium_wellbeing_countries"],Clustering_countries["high_wellbeing_countries"])
print("Statistic:",statistic)
print("P-value:",p_value)
```

```python
from scikit_posthocs import posthoc_dunn
def comparison_feature(column):
    comparison = posthoc_dunn(df_2023, val_col=column, group_col="Cluster_3", p_adjust="holm")
    return comparison
```
